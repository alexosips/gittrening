# Виконання практичної роботи

0. Встановимо необхідні бібліотеки та імпортуємо необхідні класи:

```python
!pip3 install mpmath numpy prettytable
from math import pow, tan, radians, sin, cos
from mpmath import sec
from numpy import arange
from prettytable import PrettyTable
```

1. Ознайомились з теоретичними матеріалами з виконуваної практичної роботи.

2. Варіант роботи:

```python
#Об'явимо необхідні функції та змінну

EPSILON = 0.0001 # необхідна точність

# Нехай y = f(x)

y  = lambda x: tan(1.2*x) -2 + 3*x                              # Функція
y2 = lambda x: ( 6*pow(sec(6/5*x), 2) ) / 5 +3                  # Похідна функції
y3 = lambda x: ( 72*sin(6/5*x) ) / ( 25*pow(cos(6/5*x), 3) )    # Друга похідна


def print_table(table:list, title:str) -> None:
    ''' 
    Функція "гарного" друку "таблиці" певних значень
    table:list - список значень
    title:str  - маркування рядків таблиці
    ''' 
    t = PrettyTable([title[0]]+ table[0])
    for i in range(1, len(title)):
        t.add_row([title[i]]+ [round(value, 1) for value in table[i]])
    print(t)
```

3. Виконаємо відокремлення коренів рівняння f (x) = 0  ` f(x) = tan1.2x - 2 + 3x ` на інтервалі [-5,5] з кроком 0.2

```python
table1 = [[], []]
for x in range(-5, 5+1):
    table1[0].append(x)
    table1[1].append(y(x))
print_table(table1, ['x', 'y'])
```

4. На частковому інтервалі [0;1] f змінює знак, тобто виконується `f(0)f(1)<0`. Для виконання достатні умов існування коренів ` f'(a)f'(b)>0 ` та ` f''(a)f''(b)>0 ` скануємо f з кроком 0.2

```python
table2 = [[], [], [], []]
for x in arange(0, 0.2+1, 0.2):
    table2[0].append(round(x, 1))
    table2[1].append(y(x))
    table2[2].append(y2(x))
    table2[3].append(y3(x))
print_table(table2, ['x', 'y', "y'", 'y"'])
```

Встановлюємо, що ` f'(0.4)f'(0.6)>0 ` та ` f"(0.4)f"(0.6)>0 `, тобто на частковому інтевалі [0.4;0.6] виконуються необхідні умови для існування кореня, і отже тут існує один і тільки один корінь рівння `f(x)`

5. Виконаємо уточнення кореня рівняння мотодом хорд на проміжку [0.4;0.6]. `f'(0.4)=5.1`; `f"(0.6)=4.5` -> `y'y">0`, тоді:

Процес уточнення кореня закінчується тоді, коли оцінка отриманого наближення задовольняє заданій точності ε тобто, коли абсолютна величина різниці між двома наступними наближеннями $a_{n-1} $ i $ a_{n}$ :     $| a_{n} - a_{n-1} | < \varepsilon$

```python
a_previous = 0.4
a = 0
b = 0.6
while not abs(a - a_previous) < EPSILON:
    a_previous, a = a_previous - ( y(a_previous)*(b-a_previous) ) / ( y(b) - y(a_previous) ), a_previous
print(f'Bважаємо, що коренем рівняння є х = {a:.5f}')
```

6. Виконаємо уточнення кореня рівняння методом дотичних (Ньютона).

Для вибору розрахункової формули методу дотичних визначаємо знак добутку першої та другої похідних функції f (x) в точці $x_0$ , що належить інтервалу [0,4; 0,6]. `y'y">0` , тому для уточнення кореня даного рівняння використовується формула:

```python
b_previous = 0.6
b = 0
while not abs(b - b_previous) < EPSILON:
    b_previous, b = b_previous - y(b_previous) / y2(b_previous), b_previous
print(f'Bважаємо, що коренем рівняння є х = {b:.5f}')
```
